```python
import pandas as pd
import numpy as np

# -----------------------------------------------------------
# حل مشكلة "فخ التعاونات" + تحسين دقة الأسماء
# (The Collaboration Trap + Case Insensitive Fix)
# -----------------------------------------------------------

print("🔧 جاري إصلاح قيم الفنانين (Advanced Artist Exploding)...")

# 1. إنشاء نسخة مؤقتة وتجهيزها
temp_df = df[['artists', 'popularity']].copy()
# تنظيف الأصفار (اختياري بس مستحسن)
temp_df = temp_df[temp_df['popularity'] > 0]

# 2. تفجير القوائم (Explode) لفصل الفنانين
temp_df['artists_split'] = temp_df['artists'].str.split(';')
exploded_df = temp_df.explode('artists_split')

# 3. تنظيف وتوحيد الأسماء (Lowercase + Strip)
# ده هيخلي 'Drake ' و 'DRAKE' و 'Drake' حاجة واحدة
exploded_df['artists_clean'] = exploded_df['artists_split'].str.lower().str.strip()

# # 4. حساب الوسيط الدقيق (The Accurate Map)
# accurate_artist_map = exploded_df.groupby('artists_clean')['popularity'].median().to_dict()
# 4. حساب الوسيط الدقيق (باستخدام الـ 80th Percentile للأمان)
# بنقول للموديل: قيم الفنان بناءً على مستوى "جيد جداً" من أغانيه، وتجاهل أقصى شطحاته وأدنى سقطاته
accurate_artist_map = exploded_df.groupby('artists_clean')['popularity'].quantile(0.80).to_dict()

print(f"✅ تم حساب شهرة {len(accurate_artist_map)} فنان (بمعيار الـ 80%).")

print(f"✅ تم حساب شهرة {len(accurate_artist_map)} فنان بشكل فردي دقيق.")

# 5. دالة استخراج السكور المحسنة (V2)
def get_smart_artist_score_v2(artist_string):
    if pd.isna(artist_string):
        return 45.0 # نعاقب المجهولين برقم أقل من المتوسط (50)
    
    # نفصل الأسماء في الأغنية الحالية
    artists_list = str(artist_string).split(';')
    
    scores = []
    for artist in artists_list:
        # تنظيف الاسم قبل البحث عنه (نفس منطق التدريب)
        artist_key = artist.lower().strip()
        
        # البحث في القاموس
        val = accurate_artist_map.get(artist_key)
        
        if val is None:
            # محاولة بحث بديلة (احتياطي)
            val = 45.0 
        
        scores.append(val)
    
    # نستخدم الماكسيموم بدلاً من المتوسط للنجوم الكبار!
    # المنطق: لو نجم (90) غنى مع حد مغمور (20)، الأغنية هتتشهر بسببه
    return np.max(scores) 

# 6. تطبيق الدالة وتحديث العمود في الداتا الأصلية
df['artists_encoded'] = df['artists'].apply(get_smart_artist_score_v2)

# ================== ربط الكود بباقي النوت بوك ==================

# أ) تحديث اسم القاموس العام عشان دالة التوقع (predict) تشوفه
artist_median_map = accurate_artist_map 

# ب) حذف عمود 'artists' النصي (لأننا خلاص حولناه لأرقام)
if 'artists' in df.columns:
    df.drop(columns=['artists'], inplace=True)

print("\n🔍 فحص النتائج النهائية:")
# نجرب نبحث عن 'the weeknd' بالحروف الصغيرة للتأكد
print(f"New Smart Score for 'the weeknd': {artist_median_map.get('the weeknd', 'Not Found')}")
```

### SFS (Sequential Forward Selection)

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

# ---------------------------------------------------------
# 1. تجهيز الداتا بالأسماء (عشان نعرف نختار)
# ---------------------------------------------------------
# بنجيب أسماء الأعمدة الأصلية من X (قبل الـ scaling)
feature_names = X.columns.tolist()

# بنحول مصفوفة الأرقام (Scaled) لـ DataFrame عشان يبقى ليها أسماء
X_train_df = pd.DataFrame(X_train_scaled, columns=feature_names)

# ---------------------------------------------------------
# 2. خطوة العينة (Sampling) - هنا كان حل المشكلة 🛠️
# ---------------------------------------------------------
# بناخد عينة عشوائية (5000 صف) من X و y مع بعض عشان نضمن تطابقهم
# لاحظ: y_train لازم تكون متوافقة، فبنعملها reset_index احتياطي داخل دالة الـ split
X_sample, _, y_sample, _ = train_test_split(
    X_train_df, 
    y_train.reset_index(drop=True), # حل مشكلة الـ Index هنا
    train_size=5000, 
    random_state=42
)

print(f"تم أخذ عينة حجمها: {X_sample.shape} لتسريع عملية الاختيار.")

# ---------------------------------------------------------
# 3. إعداد الموديل والـ SFS
# ---------------------------------------------------------
# موديل خفيف وسريع (عدد أشجار قليل وعمق محدود)
rf_fast = RandomForestRegressor(n_estimators=20, max_depth=5, n_jobs=-1, random_state=42)

# إعداد خوارزمية الاختيار المتسلسل
sfs = SequentialFeatureSelector(
    rf_fast, 
    n_features_to_select=15,  # نقيلي أهم 15 أعمدة (ممكن تغير الرقم ده)
    direction='forward',      # ابني الفريق واحد واحد
    scoring='r2',             # معيار الجودة
    cv=3                      # جرب 3 مرات للتأكيد
)

print("جاري اختيار أفضل Features... (ممكن ياخد دقيقة أو دقيقتين ☕)")

# ---------------------------------------------------------
# 4. التشغيل واستخراج النتائج
# ---------------------------------------------------------
sfs.fit(X_sample, y_sample)

# استخراج أسماء الأعمدة المختارة
selected_mask = sfs.get_support()
selected_features = np.array(feature_names)[selected_mask]

print("-" * 40)
print(f"🎉 تم اختيار أفضل {len(selected_features)} ميزات وهم:")
print(selected_features)
print("-" * 40)

# ---------------------------------------------------------
# 5. الفلترة النهائية (عشان تكمل بيها الموديل)
# ---------------------------------------------------------
# بنحتفظ بس بالأعمدة دي في داتا التدريب والاختبار
X_train_final = X_train_df[selected_features]

# لازم نحول X_test_scaled لـ DataFrame بنفس الطريقة عشان نختار منه نفس الأعمدة
X_test_final = pd.DataFrame(X_test_scaled, columns=feature_names)[selected_features]

print("✅ تم تحديث X_train_final و X_test_final بالأعمدة المختارة فقط.")
```

```python
import xgboost as xgb
import optuna
from sklearn.metrics import mean_squared_error

# دالة الهدف المحسنة (Aggressive Mode) للحصول على أرقام أعلى وأكثر واقعية للنجوم
def objective_aggressive(trial):
    # تعريف الـ Hyperparameters الهجومية (تسمح للموديل بالتعلم بشكل أعمق وأسرع)
    params = {
        'objective': 'reg:squarederror',
        'n_estimators': 600,  # عدد أشجار أكبر
        # زيادة معدل التعلم لتسريع التقاط الأنماط القوية
        'learning_rate': trial.suggest_float('learning_rate', 0.05, 0.2), 
        # السماح بعمق أكبر للشجرة لفهم التفاصيل الدقيقة للأغاني الناجحة
        'max_depth': trial.suggest_int('max_depth', 6, 12), 
        # السماح بمجموعات أصغر (يساعد في التقاط الأغاني الفريدة)
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 5), 
        'subsample': trial.suggest_float('subsample', 0.7, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.7, 1.0),
        # تقليل القيود (Regularization) بشكل كبير لكي لا يكبت الأرقام العالية
        'reg_alpha': trial.suggest_float('reg_alpha', 0, 2), 
        'reg_lambda': trial.suggest_float('reg_lambda', 0, 2),
        'n_jobs': -1
    }
    
    # تدريب الموديل بالمواصفات المقترحة
    model = xgb.XGBRegressor(**params)
    model.fit(X_train_final, y_train)
    
    # التقييم
    preds = model.predict(X_test_final)
    rmse = mean_squared_error(y_test, preds, squared=False)
    return rmse

print("🚀 Starting Aggressive Bayesian Optimization using Optuna...")
# إنشاء دراسة جديدة
study = optuna.create_study(direction='minimize') 
# تجربة 15 محاولة (كافية مع هذه الإعدادات القوية)
study.optimize(objective_aggressive, n_trials=15) 

print("\n✅ Best Hyperparameters found:")
print(study.best_params)

# تدريب الموديل النهائي بأفضل أرقام وصلنالها
print("\n🏋️ Training Final Aggressive XGBoost Model...")
best_xgb = xgb.XGBRegressor(**study.best_params, n_estimators=600, n_jobs=-1)
best_xgb.fit(X_train_final, y_train)
print("✅ Done. Model is ready to predict HITS!")
```

```python
import pandas as pd
import numpy as np

def predict_song_success_fixed(song_details):
    # 1. إنشاء DataFrame من المدخلات
    input_df = pd.DataFrame([song_details])
    
    # -------------------------------------------------------
    # 2. هندسة الخصائص (Feature Engineering)
    # -------------------------------------------------------
    # التأكد من عدم القسمة على صفر
    danceability = input_df['danceability'].iloc[0] if input_df['danceability'].iloc[0] > 0 else 0.001
    
    input_df['duration_min'] = input_df['duration_ms'] / 60000
    input_df['energy_dance_ratio'] = input_df['energy'] / danceability
    input_df['electronic_score'] = 1 - input_df['acousticness']
    
    # -------------------------------------------------------
    # 3. التحويل الرقمي (Encoding) - التحديث الأخير 🚀
    # -------------------------------------------------------
    artist_text = song_details.get('artists')
    genre_text = song_details.get('track_genre')
    
    # --- معالجة الفنانين (Advanced) ---
    if pd.isna(artist_text):
        input_df['artists_encoded'] = 45.0 # عقاب للمجهولين
    else:
        # فصل الأسماء
        artists_list = str(artist_text).split(';')
        scores = []
        for a in artists_list:
            # تنظيف وتوحيد حالة الأحرف (Lowercase)
            key = a.lower().strip()
            # البحث في القاموس الدقيق، ولو مش موجود نستخدم 45
            scores.append(accurate_artist_map.get(key, 45.0))
        
        # استخدام الماكسيموم: قوة النجم الأكبر هي اللي بتحدد نجاح الأغنية
        input_df['artists_encoded'] = np.max(scores)
        
    # --- معالجة النوع (Genre) ---
    default_val = df['popularity'].median() if 'popularity' in df else 50.0
    
    if genre_text in median_map:
        input_df['track_genre'] = median_map[genre_text]
    else:
        input_df['track_genre'] = default_val

    # -------------------------------------------------------
    # 4. الترتيب والتحجيم (Scaling & Ordering)
    # -------------------------------------------------------
    training_columns = X.columns.tolist() 
    
    for col in training_columns:
        if col not in input_df.columns:
            input_df[col] = 0
            
    input_df = input_df[training_columns]
    input_scaled_array = scaler.transform(input_df)
    input_scaled_df = pd.DataFrame(input_scaled_array, columns=training_columns)
    
    # -------------------------------------------------------
    # 5. التوقع (Prediction)
    # -------------------------------------------------------
    final_input = input_scaled_df[selected_features]
    
    pred_xgb = best_xgb.predict(final_input)[0]
    pred_rf = rf_model.predict(final_input)[0]
    pred_nn = model.predict(final_input, verbose=0).flatten()[0]
    
    # طباعة النتائج الفردية (Debug)
    print(f"🔍 Debug Info:")
    print(f"   - Encoded Artist Value: {input_df['artists_encoded'].values[0]}")
    print(f"   - Encoded Genre Value: {input_df['track_genre'].values[0]}")
    print(f"   - XGBoost: {pred_xgb:.2f}")
    print(f"   - Random Forest: {pred_rf:.2f}")
    print(f"   - Neural Net: {pred_nn:.2f}")
    
    # الدمج النهائي (تعديل الأوزان لصالح RF قليلاً لأنه كان مستقراً)
    final_score = (pred_xgb * 0.70) + (pred_rf * 0.20) + (pred_nn * 0.10)
    
    # تقييد النتيجة
    final_score = max(0, min(100, final_score))
    
    return final_score

# --- اختبار الكود ---
hit_song = {
    'artists': 'Ed Sheeran',       
    'track_genre': 'pop',          
    'duration_ms': 210000,
    'danceability': 0.85,
    'energy': 0.80,
    'loudness': -5.0,
    'acousticness': 0.10,
    'speechiness': 0.05,
    'liveness': 0.10,
    'valence': 0.90,
    'tempo': 120,
    'key': 1,
    'mode': 1,
    'time_signature': 4,
    'explicit': 0,
    'instrumentalness': 0.0
}

score = predict_song_success_fixed(hit_song)
print(f"\n🎵 النتيجة النهائية المتوقعة: {score:.2f} / 100")
```
